```python
from moviepy.editor import *
import os
import shutil
import pygame
```

```python
if os.path.exists("Videos"):
    shutil.rmtree("Videos")
os.mkdir("Videos")
```

```python
os.listdir() #Your files should be included in the same directory as the script
```

```python
for file in os.listdir():
    if file.endswith(".mp4"):
        shutil.copy(file, f"Videos/{file}")
    if len(os.listdir("Videos")) >= 9:
        break
```

```python
os.listdir("Videos")
```

```python
cliplist = {}
n = 1
for file in os.listdir("Videos"):
    cliplist[n] = VideoFileClip(f"Videos/{file}").fx(afx.audio_normalize).resize(height=180, width=320)
    n+=1
```

```python
#cliplist[1].ipython_display(width=280)
```

```python
while n < 10: #Code to create extra black screens
    cliplist[n] = ColorClip(size = (180, 320), color = [0, 0, 0], duration = 1)
    n +=1
```

```python
cliplist
```

```python
finalclip = clips_array([[cliplist[1], cliplist[2], cliplist[3]],
                         [cliplist[4], cliplist[5], cliplist[6]],
                         [cliplist[7], cliplist[8], cliplist[9]]])
```

```python
finalclip.resize(width=100)
finalclip.write_videofile("Results/CompositeClip3.mp4", codec = 'libx264',
                                               audio_codec='aac',
                                               temp_audiofile = 'Results/temp_audio.m4a', 
                                               remove_temp = True
                                               )
```
